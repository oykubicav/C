```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os # Import the os module

input_dir="Star Datas/Star Datas" # Corrected input directory path
output_dir = "ProcessedData"
os.makedirs(output_dir, exist_ok=True)

# Correctly call sorted with the list comprehension
image_filenames = sorted([
    f for f in os.listdir(input_dir) if f.lower().endswith(('.jpg', '.png'))
])

def preprocess_star_image(img_path):
  img = cv2.imread(img_path,cv2.IMREAD_GRAYSCALE)
  clahe = cv2.createCLAHE(clipLimit=2.0,tileGridSize=(8,8))
  img_clahe = clahe.apply(img)

  if img_clahe.dtype != np.uint8:
    img_clahe = cv2.convertScaleAbs(img_clahe)
  img_blur = cv2.GaussianBlur(img_clahe,(5,5),0)

  _, img_thresh = cv2.threshold(img_blur, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
  return img_thresh

for filename in image_filenames:
    in_path = os.path.join(input_dir, filename)
    out_path = os.path.join(output_dir, filename)

    processed = preprocess_star_image(in_path)
    cv2.imwrite(out_path, processed)
```

```python
import cv2
import os
from glob import glob
class_names =  []


detect_dir = "runs/detect/constellation_test"
label_dir = os.path.join(detect_dir, "labels")
output_dir = "overlayed"
os.makedirs(output_dir, exist_ok=True)

image_paths = glob(os.path.join(detect_dir, "*.jpg"))

def draw_predictions(img_path, label_path, out_path):
    img = cv2.imread(img_path)
    h, w, _ = img.shape
    if not os.path.exists(label_path):
        print(f"No labels for {img_path}")
        return
    with open(label_path, 'r') as f:
        for line in f.readlines():
            cls, x, y, bw, bh, conf = map(float, line.strip().split())
            x1 = int((x - bw / 2) * w)
            y1 = int((y - bh / 2) * h)
            x2 = int((x + bw / 2) * w)
            y2 = int((y + bh / 2) * h)
            cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(img, class_names[int(cls)], (x1, y1 - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 1)
            cv2.putText(img,)

for img_path in image_paths:
    filename = os.path.basename(img_path)
    label_path = os.path.join(label_dir, filename.replace(".jpg", ".txt"))
    out_path = os.path.join(output_dir, filename)
    draw_predictions(img_path, label_path, out_path)

print(f"✅ {len(image_paths)} görsel çizildi. Sonuçlar: /overlayed/")
```

```python
import os
import json
from glob import glob

# Class names from stars.yaml
class_names = ['aquila', 'bootes', 'canis_major', 'canis_minor', 'cassiopeia', 'cygnus', 'gemini', 'leo',
               'lyra', 'moon', 'orion', 'pleiades', 'sagittarius', 'scorpius', 'taurus', 'ursa_major']

# Paths
img_dir = "runs/detect/constellation_test"
label_dir = os.path.join(img_dir, "labels")
output_json_path = "constellation_detections.json"

all_detections = []

# Read each label file
for label_file in glob(os.path.join(label_dir, "*.txt")):
    image_name = os.path.basename(label_file).replace(".txt", ".jpg")
    image_path = os.path.join(img_dir, image_name)
    detections = []

    with open(label_file, "r") as f:
        for line in f:
            parts = line.strip().split()
            cls_id = int(parts[0])
            cx, cy, bw, bh = map(float, parts[1:5])
            conf = float(parts[5]) if len(parts) == 6 else None  # if confidence exists
            detections.append({
                "class": class_names[cls_id],
                "class_id": cls_id,
                "confidence": conf,
                "bbox_normalized": [cx, cy, bw, bh]
            })

    all_detections.append({
        "image": image_name,
        "image_path": image_path,
        "detections": detections
    })

# Save all in a single JSON file
with open(output_json_path, "w") as f:
    json.dump(all_detections, f, indent=2)

print(f"✅ Detection results saved to: {output_json_path}")
```
